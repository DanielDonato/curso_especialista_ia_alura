```python
import pandas as pd
```

```python
data = pd.read_csv('./data/resenhas.csv')
```

```python
data.head()
```

```python
results_predict = sentiment_model_classifier.predict(data['Resenha'].astype(str).tolist())
results_predict
```

```python
sentiments = []

for result in results_predict:
    sentiments.append(result['label'])

sentiments
```

```python
data['Sentimento'] = sentiments
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

portuguese_stopword = set(stopwords.words('portuguese'))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def word_cloud(texto, column_text, sentiment):
    text_sentiment = texto.query(f"Sentimento == '{sentiment}'")[column_text]

    text_join = " ".join(text_sentiment)

    words = text_join.split()
    filter_words = [word for word in words if word not in portuguese_stopword]
    filter_words = " ".join(filter_words)

    word_cloud = WordCloud(width=800, height=500, max_words=50).generate(filter_words)
    plt.figure(figsize=(10, 7))
    plt.imshow(word_cloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
word_cloud(data, 'Resenha', 'POS')
```

```python
word_cloud(data, 'Resenha', 'NEG')
```
