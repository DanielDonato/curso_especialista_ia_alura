```python
import pandas as pd
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli",
                      device=-1)
```

```python
categories = ['eletronics', 'food', 'toys', 'books']
```

```python
predict = classifier(""""
Latest model of smartphone with 5G connectivity and 128GB internal storage
""", candidate_labels=categories)
```

```python
data = pd.read_csv('./data/descricoes_produtos.csv')
data
```

```python
candidate_categories = ['eletrodomésticos', 'eletrônicos', 'beleza', 'brinquedos']

def categorize(_description):
    _result = classifier(_description, candidate_categories, multi_label=False)
    category = max(zip(_result['labels'], _result['scores']), key=lambda x: x[1])[0]
    return category
```

```python
data['Categoria'] = data['Descrição'].apply(categorize)
```
