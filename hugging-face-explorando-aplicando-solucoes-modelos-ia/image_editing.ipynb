```python
import os

os.environ["CUDA_VISIBLE_DEVICES"] = ""

from transformers import pipeline
from transformers.modeling_utils import PreTrainedModel

# Compatibility fix for some older/custom Hugging Face remote models
# that do not define all_tied_weights_keys but newer transformers expects it.
if not hasattr(PreTrainedModel, "all_tied_weights_keys"):
    PreTrainedModel.all_tied_weights_keys = {}

image_path = "https://farm5.staticflickr.com/4007/4322154488_997e69e4cf_z.jpg"
pipe = pipeline("image-segmentation", model="briaai/RMBG-1.4", trust_remote_code=True, device=-1)
pillow_mask = pipe(image_path, return_mask = True) # outputs a pillow mask
pillow_image = pipe(image_path) # applies mask on input and returns a pillow image
```

```python
import gradio as gr
from transformers import pipeline
from PIL import Image

def remove_background(image):
    _pipe = pipeline("image-segmentation", model="briaai/RMBG-1.4", trust_remote_code=True)

    _pillow_mask = _pipe(image, return_mask=True)

    _pillow_image = _pipe(image)

    return _pillow_image
```

```python
app = gr.Interface(
    fn=remove_background,
    inputs=gr.components.Image(type="pil"),
    outputs=gr.components.Image(type="pil", format='png'),
    title="Remove Background",
    description="Envie uma imagem e veja o background sendo removido automaticamente. A imagem resultante será no formato PNG."
)
app.launch()
```
