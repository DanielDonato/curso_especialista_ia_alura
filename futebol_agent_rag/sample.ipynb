```python
# ImportaГ§Гµes bГЎsicas
import os

# loader de documentos PDF
from langchain_community.document_loaders import PyPDFLoader

# DivisГЈo de texto em blocos
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Embeddings
from langchain_community.embeddings import HuggingFaceEmbeddings

# Banco vetorial
from langchain_community.vectorstores import Chroma

# LLM
from langchain_openai import ChatOpenAI

# Cadeia RAG
from langchain.chains import RetrievalQA

from langchain_groq import ChatGroq
```

```python
from dotenv import load_dotenv
load_dotenv()

groq_api_key = os.environ.get('API_KEY')
```

```python
pdf_path = "./data/regras_futebol.pdf"

loader = PyPDFLoader(pdf_path)
documents = loader.load()
open_ai_api_key = os.environ.get('API_KEY')
len(documents)
```

```python
# Divide os documentos em chunks menores
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

chunks = text_splitter.split_documents(documents)

len(chunks)
```

```python

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
# Cria o banco vetorial
vectorstore = Chroma.from_documents(
    documents=chunks,
    embedding=embeddings,
    persist_directory="./models/chroma_regras_futebol"
)
```

```python
# Cria o retriever
retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 3}
)
```

```python
# Inicializa o modelo de linguagem
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    groq_api_key=groq_api_key
)
```

```python
# Cria a cadeia RAG
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True)
```

```python
# Pergunta de teste
pergunta = "Um jogador pode usar a mГЈo para marcar um gol"

# Executa a pergunta no agente RAG
resposta = qa_chain(pergunta)

print("Pergunta:")
print(pergunta)

print("\nResposta do Agente:")
print(resposta["result"])

print("\nTrechos utilizados como contexto:\n")
for i, doc in enumerate(resposta["source_documents"], start=1):
    print(f"--- Trecho {i} ---")
    print(f"Fonte: {doc.metadata.get('source', 'Documento desconhecido')}")
    print(f"PГЎgina: {doc.metadata.get('page', 'N/A')}\n")
    print("ConteГєdo:")
    print(doc.page_content)
    print("\n")
```
